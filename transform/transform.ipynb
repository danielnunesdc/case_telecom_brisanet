```python
import pandas as pd
import re
from unidecode import unidecode
```

```python
df_clientes = pd.read_csv(r"../input/base_clientes.csv", delimiter=";", encoding="utf-8")
df_produtos = pd.read_csv(r"../input/base_produtos.csv", delimiter=";", encoding="utf-8")
df_vendas = pd.read_csv(r"../input/base_vendas.csv", delimiter=";", encoding="utf-8")
df_info_vendas = pd.read_csv(r"../input/Informacoes_vendas.csv", delimiter=";", encoding="utf-8")
```

#### CLIENTES

```python
df_clientes.head(10)
```

```python
# Primeiro passo normalziar o df, remover acentos e carc. esp. que não quero
def normalizar_para_snake_case(valores):

    if pd.isna(valores):
        return valores 

    valores = str(valores).strip()
    valores = unidecode(valores)
    valores = valores.lower()

    valores = re.sub(r"[^a-z0-9\+\-\s]+", " ", valores)

    valores = re.sub(r"\s+", " ", valores).strip()

    # Palavras que NÃO devem ser capitalizadas
    excecoes = {"de", "do", "da"}

    palavras = []
    for palavra in valores.split(" "):
        if palavra in excecoes:
            palavras.append(palavra)
        else:
            palavras.append(palavra.capitalize())

    valores = " ".join(palavras)

    return valores
```

```python
colunas = ['cidade', 'segmento', 'plano']

for col in colunas:
    df_clientes[col] = df_clientes[col].apply(normalizar_para_snake_case)
```

```python
df_clientes.head(10)
```

```python
# Normalizar estado

def normalizar_estado(valor):
    if pd.isna(valor):
        return valor

    valor = unidecode(str(valor)).strip().upper()
    valor = re.sub(r"[^\w]+", "", valor)
    return valor

df_clientes["estado"] = df_clientes["estado"].apply(normalizar_estado)
```
